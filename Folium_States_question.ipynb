```
import pandas as pd
```

```
import folium
```

```

db = "https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv"
map_data = "https://raw.githubusercontent.com/KiranJoshi25/DataVizFall2020_Colab-/master/cb_2018_us_state_20m.json"
df = pd.read_csv(db)
df.head(5)
```

```
map_data = "https://raw.githubusercontent.com/KiranJoshi25/DataVizFall2020_Colab-/master/cb_2018_us_state_20m.json"
state = pd.read_json(map_data)
state['features'].iloc[0]
```

```
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

```
def covid_stat_on(df,strDate):
  df_temp = df[df['date']==strDate][['fips','state','cases']].copy().dropna().reset_index(drop=True)
  #df_temp['fips'] = df_temp['fips'].astype('int').astype('str').apply(lambda x: x.zfill(5))
  return df_temp
covid_stat_on(df,'2020-9-13')
```

```
#map_data = "https://raw.githubusercontent.com/KiranJoshi25/DataVizFall2020_Colab-/master/cb_2018_us_state_20m.json"
m = folium.Map(
            location=[48,-102],
            zoom_start=4)

folium.Choropleth(
    geo_data=map_data, 
    name='Covid Cases by state',               
    legend_name='Actual Covid Cases',  
    data=covid_stat_on(df,'2020-9-13'),  
    columns=['state','cases'],             
    key_on='feature.properties.NAME',
    fill_color='RdPu',                         
    fill_opacity=0.7,
    line_opacity=0.2,
    nan_fill_color='RED',
    highlight=True,
).add_to(m)

m
```

```
m = folium.Map(
            location=[48,-102],
            zoom_start=4)
folium.Choropleth(
    geo_data=map_data, 
    name='Covid Cases by state',               
    legend_name='Magnitude',  
    data=df_mag,  
    columns=['state','cases'],             
    key_on='feature.properties.NAME',
    fill_color='BuGn',                         
    fill_opacity=0.7,
    line_opacity=0.2,
    nan_fill_color='None',
    highlight=True,
).add_to(m)

folium.Choropleth(
    geo_data=map_data, 
    name='Covid Cases by state',               
    legend_name='Actual Covid Cases',  
    data=covid_stat_on(df,'2020-9-13'),  
    columns=['state','cases'],             
    key_on='feature.properties.NAME',
    fill_color='RdPu',                         
    fill_opacity=0.7,
    line_opacity=0.2,
    nan_fill_color='blue',
    highlight=True,
).add_to(m)
folium.LayerControl().add_to(m)
m
```
